```python
def getSummary(link, contains_images = "No"):
    try:
        summaryPage = urllib.request.urlopen(link)
        summarySoup = BeautifulSoup(summaryPage, 'html.parser')
        summary = summarySoup.find("tbody").find_all("tr")[1].find("td").text
        img_urls = []
        links = summarySoup.find_all("img", {"class": "ufopic"})
        for link in links: 
            img_urls.append(link["src"])
        return summary, img_urls
    except Exception:
        pass
```

```python
from datetime import date
from datetime import datetime
import time
import pandas as pd
import urllib.request
import threading
from multiprocessing.pool import ThreadPool as pool
from bs4 import BeautifulSoup
import time

import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
allUrl = "http://www.nuforc.org/webreports/ndxevent.html"
allPage = urllib.request.urlopen(allUrl)
allSoup = BeautifulSoup(allPage, 'html.parser')

table = allSoup.find("table")

linksByDate = []
for link in table.find_all("a"):
    linksByDate.append(link['href'])
```

```python
f = open("console_output.txt", "a")
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
f.write("file open at =" + current_time + "\n")

pool_size = 3
pool = pool(pool_size)

now = datetime.now()
current_time = now.strftime("%H:%M:%S")
f.write("pool created at =" + current_time + " and pool_size =" + str(pool_size) +"\n")
```

```python
count = 0
monthRows = []
key = 0

now = datetime.now()
current_time = now.strftime("%H:%M:%S")
f.write("collection begin at =" + current_time + "\n")

for month in linksByDate:
    monthUrl = 'http://www.nuforc.org/webreports/' + month
    time.sleep(2.5)
    monthPage = urllib.request.urlopen(monthUrl)
    monthSoup = BeautifulSoup(monthPage, 'html.parser')
    monthTable = monthSoup.find("tbody")
    
    now = datetime.now()
    current_time = now.strftime("%H:%M:%S")
    f.write("read monthUrl= " + monthUrl + " at =" + current_time + "\n")
    
    if(monthUrl != "http://www.nuforc.org/webreports/ndxe.html"):
        for x in monthTable.find_all("tr"):
            link = x.find("a")
            if(link is not None):
                pool.apply_async(report_worker, (link, key,))
                key += 1
    now = datetime.now()

now = datetime.now()
current_time = now.strftime("%H:%M:%S")
f.write("pool.close() at =" + current_time + "\n")

pool.close()

now = datetime.now()
current_time = now.strftime("%H:%M:%S")
f.write("pool.join() called at =" + current_time + "\n")

pool.join()

now = datetime.now()
current_time = now.strftime("%H:%M:%S")
f.write("pool.join() finished at =" + current_time + "\n")
```
